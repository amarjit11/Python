#### Variables are fundamental elements in a programming lanaguage used to store data which can be referenced or manipulated as per need

```python
## type conversion in python- python provides built-in functions for converting between different data types. for example, we can use the int() function to convert a string to an integer, and the str() function to convert an integer to a string. it is important to understand the type conversion rules of python in order to write correct and efficient code. for example, the following code will produce the expected result because it follows the correct semantics of python.

age=25
print(type(age))
age_str=str(age)
print(age_str)
print(type(age_str))
```

```python
height=5.11
print(type(height))
height_int=int(height)
print(height_int)
print(type(height_int))
height_int=float(int(height_int))
print(type(height_int))
```

```python
## dynamic typing in python- python is a dynamically typed language, which means that the type of a variable is determined at runtime. this allows for greater flexibility in coding, but it also means that we need to be careful when working with variables to avoid type errors. for example, the following code will produce a type error because we are trying to add an integer and a string.

x=10
print(x,type(x))
x="amarjit"
print(x,type(x))
x=9.12
print(x,type(x))
```

```python
## input function in python- the input() function is used to take input from the user. it returns a string, so we need to use type conversion to convert the input to the desired data type. for example, the following code will take an integer input from the user and print it.

agee=str(input("enter your age: "))
age_int=int(agee)
print("your age is:",age_int,type(age_int))
```

```python
## simple calculator- the following code is a simple calculator that takes two numbers and an operator as input from the user and performs the desired operation.

num1=float(input("enter first number: "))
num2=float(input("enter second number: "))
operator=input("enter operator (+, -, *, /): ")

if operator=='+':
	print("result:",num1+num2)
elif operator=='-':
	print("result:",num1-num2)
elif operator=='*':
	print("result:",num1*num2)
elif operator=='/':
	if num2!=0:
		print("result:",num1/num2)
	else:
		print("error: division by zero")
else:
	print("error: invalid operator")		
```
