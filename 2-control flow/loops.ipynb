```python
range(0,5)
```

```python
## for loop

for i in range(1,6):
	print(i)
```

```python
for i in range(11,1,-2):
	print(i)
```

```python
# strings
str="Amarjit"
for i in str:
	print(i)
```

```python
## while loop continue to execute the loop until the conditions are true

count=0
while count<5:
	print(count)
	count+=1
```

```python
## loop control statement
## break statement exits the loop prematurely

for i in range(10):
	if i==5:
		break
	if i !=5:
		print(i)
```

```python
## continue skips the cuurent iteration and continue with next iteration

for i in range(10):
	if i%2==0:
		continue
	print(i)


```

```python
## pass is null statement it does nothing
for i in range(5):
	if i==3:
		pass
	print(i)
```

```python
## nested loop a loop inside a loop

for i in range(3):
	for j in range(2):
		print(f"i:{i} and j:{j}")
```

```python
## calculate sum of first n natural number using while loop

n= int(input("enter number"))
sum=0
count=1
while count<=n:
	sum=sum+count
	count+=1
print(sum)
```

```python
## for loop

n=int(input())
sum=0
count=1
for i in range(n):
	sum+=count
	count+=1
print(sum)
```

```python
## prime numbers between 1 to 100

for num in range(101):
	if num>1:
		for i in range(2,num):
			if num%i==0:
				break
		else:
			print(num)

			
```
