#### Conditional Statements

```python
## conditional statements
## if statement
x = 10
if x > 5:
	print("x is greater than 5")
```

```python
x>=9
```

```python
## else statement
x = 3
if x>5:
	print("x is greater than 5")
else:
	print("x is less than 5")
```

```python
## elif statement
age=20
if age<18:
	print("you are a minor")
elif age>=18 and age<60:
	print("you are an adult")
else:
	print("you are a senior citizen")
```

```python
## nested conditional statements
num=int(input("enter the number"))
if num>=0:
	if num==0:
		print("the number is zero")
	if num%2==0:
	    print("the number is even")
	else:
		print("the number is odd")
else:
		print("the number is negative")	
```

```python
## practical examples
## determing a year is leap year using nested conditional statements
year= int(input("enter the year"))
if year%4==0:
	if year%100==0:
		if year%400==0:
			print(year,"the year is a leap year")
		else:
			print(year,"the year is not a leap year")
	else:
		print(year,"the year is a leap year")
```

```python
## simple calculator using nested conditional statements
x=float(input("enter first number: "))
y=float(input("enter second number: "))
operation=input("enter operation(+,-,*,/)")
if operation == '+':
	result= x+y
elif operation =='-':
	result=x-y
elif operation=='*':
	result= x*y
elif operation=='/':
	if y!=0:
		result=x/y
	else:
		print("undefined , cant be divided by 0")
else:
	print("invalid operator")
print(f"Result = ",round(result,2))
```

```python
## determine the ticket price based on age using nested conditional statements
age=int(input("enter your age"))
student=input("are you a student").lower()
ticket=100
if age<=5:
	price=ticket-(100 * ticket/100)
elif 5<age<=20:
	if student =='yes':
		price=ticket-(50*ticket/100)
	else:
		price=ticket-(80*ticket/100)
elif age>20:
	price=ticket
print(price)
```
